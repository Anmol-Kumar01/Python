Sorting in data science refers to the process of arranging or organizing data in a specific order or sequence based on one or more criteria. It is a fundamental operation in data analysis and data manipulation tasks, as it allows for efficient data retrieval, aggregation, and analysis.

```python
import random
import matplotlib.pyplot as plt
```

```python
# Global veriables to calculate the number of swaps and comaprison
quick_count_compare = 0
quick_count_swap = 0


select_count_compare = 0
select_count_swap = 0

insertion_count_compare = 0
insertion_count_swap = 0
```

```python
############################# INTERTION SORT ##################################
def insertion_sort(b):
	i = 0
	while i < len(b):
		push_down(b,i)
		i = i +1

def push_down(b, i):
	j = i 
	while j > 0 and b[j-1] > b[j]:
		global insertion_count_compare
		insertion_count_compare += 1
		insertion_swap(b, j-1, j)
		j = j-1


def insertion_swap(b, h, k):
  global insertion_count_swap
  insertion_count_swap += 1
  temp = b[h]
  b[h] = b[k]
  b[k] = temp
```

```python
############################## SELECTION SORT #############################
def select_swap(lst, x, y):
	global select_count_swap
	select_count_swap +=1
	t = lst[y]
	lst[y] = lst[x]
	lst[x] = t

def find_min(lst, i):
	min_index = i
	for j in range(i, len(lst)):
		global select_count_compare
		select_count_compare += 1
		if lst[min_index] > lst[j]:
			min_index = j
		j = j + 1
	return min_index

def select_sort(lst):
	n = len(lst)
	i = 0
	while i < n:
		minimum = find_min(lst, i)
		select_swap(lst, minimum, i)
		i = i + 1		
```

```python
############################## QUICK SORT #############################
def partition(b, h, k):
	global quick_count_compare
	i = h
	j = k+1
	x = b[h]
	while i < j-1:
		quick_count_compare +=1
		if b[i+1] >= x:
			
			quick_swap(b,i+1,j-1)
			j = j - 1
		else: # b[i+1] < x
	
			quick_swap(b,i,i+1)
			i = i + 1
	return i

def quick_swap(lst, x, y):
	global quick_count_swap
	quick_count_swap += 1 
	t = lst[x]
	lst[x] = lst[y]
	lst[y] = t

def quick_sort(b, h, k):
	if k-h < 1:
		return None
	index = partition(b, h, k)
	quick_sort(b, h, index-1)
	quick_sort(b, index+1, k) 
```

```python
# Accumulators for storing the values
x_list = []

insertion_sort_swap_list = []
selection_sort_swap_list = []
quick_sort_swap_list = []

insertion_sort_compare_list = []
selection_sort_compare_list = []
quick_sort_compare_list = []

# for loop for computing the values and appending them into their respective lists
for x in range(50, 400, 10):
	x_list.append(x)
	for i in range(3):
		a = [random.randint(1,x) for j in range(x)]
		b = a
		c = a
		insertion_sort(a)
		select_sort(b)
		quick_sort(c, 0, (len(c)-1))

# Taking averages of swaps and camparison
	insertion_sort_swap_list.append(insertion_count_swap/3)
	selection_sort_swap_list.append(select_count_swap/3)
	quick_sort_swap_list.append(quick_count_swap/3)


	insertion_sort_compare_list.append(insertion_count_compare/3)
	selection_sort_compare_list.append(select_count_compare/3)
	quick_sort_compare_list.append(quick_count_compare/3)

# Neutralise the veriables 
	quick_count_compare = 0
	quick_count_swap = 0

	select_count_compare = 0
	select_count_swap = 0

	insertion_count_compare = 0
	insertion_count_swap = 0
```

```python
# Ploting Graph for swaping
plt.plot(x_list,insertion_sort_swap_list, label = "insertion_swap" )
plt.legend()
plt.plot(x_list, selection_sort_swap_list,  label="selection_swap")
plt.legend()
plt.plot(x_list, quick_sort_swap_list,  label="quick_swap")
plt.legend()


```

```python
# Ploting Graph for comparision
plt.plot(x_list,insertion_sort_compare_list, label = "insertion_comparison" )
plt.legend()
plt.plot(x_list, selection_sort_compare_list,  label="selection_comparison")
plt.legend()
plt.plot(x_list, quick_sort_compare_list,  label="quick_comparison")
plt.legend()
```
